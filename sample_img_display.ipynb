```python
import numpy as np
import  matplotlib.pyplot as plt
from PIL import Image
```

```python
def load_face(idx):
    return np.asarray(Image.open(f'datasets\\trainset\\faces\\face{idx}.png').convert('L'))
fig, axs = plt.subplots(nrows=1, ncols=8, figsize=(16, 2))
for idx in range(len(axs)):
    axs[idx].imshow(load_face(idx+1), cmap='gray')
    axs[idx].set_xticks([])
    axs[idx].set_yticks([])
plt.suptitle('face sample for training')
plt.show()
```

```python
def load_non_face(idx):
    return np.asarray(Image.open(f'datasets\\trainset\\non-faces\\non_face{idx}.png').convert('L'))
fig, axs = plt.subplots(nrows=1, ncols=8, figsize=(16, 2))
for idx in range(len(axs)):
    axs[idx].imshow(load_non_face(idx+1), cmap='gray')
    axs[idx].set_xticks([])
    axs[idx].set_yticks([])
plt.suptitle('Non-face sample for training')
plt.show()
```

```python
basic_haar_features = []
basic_haar_features.append(np.array([[1], [0]]))
basic_haar_features.append(np.array([[0, 1]]))
basic_haar_features.append(np.array([[1], [0], [1]]))
basic_haar_features.append(np.array([[1, 0, 1]]))
basic_haar_features.append(np.array([[1, 0], [0, 1]]))
fig, axs = plt.subplots(nrows=1, ncols=5, figsize=(10, 2))
for idx in range(len(axs)):
    axs[idx].imshow(basic_haar_features[idx], cmap='gray', aspect='auto')
    axs[idx].set_xticks([])
    axs[idx].set_yticks([])
plt.suptitle('Basic haar features')
plt.show()
```
